```python
import pandas as pd
from bigquery_management.funcoes_gestao_bigquery import submeter_bigquery

from config_paths import PATH_RAW_FILES
from config_tables import TABLES

from config_bigquery import (
     RAW_TABLE_CUSTOMERS,
     RAW_TABLE_GEOLOCATION,
     RAW_TABLE_ORDER_ITEMS,
     RAW_TABLE_ORDER_PAYMENTS,
     RAW_TABLE_ORDER_REVIEWS,
     RAW_TABLE_ORDERS,
     RAW_TABLE_PRODUCTS,
     RAW_TABLE_SELLERS,
     RAW_TABLE_CATEGORY_NAME_TRANSLATION
)

for nome, (arquivo, encoding) in TABLES.items():
    
    df = pd.read_csv(PATH_RAW_FILES / arquivo, encoding=encoding)
    print(f"{nome}: {df.shape[0]} linhas, {df.shape[1]} colunas")
    
    if nome == "customers":
        submeter_bigquery(df, RAW_TABLE_CUSTOMERS)
    elif nome == "geolocation":
        submeter_bigquery(df, RAW_TABLE_GEOLOCATION)
    elif nome == "order_items":
        submeter_bigquery(df, RAW_TABLE_ORDER_ITEMS)
    elif nome == "order_payments":
        submeter_bigquery(df, RAW_TABLE_ORDER_PAYMENTS)
    elif nome == "order_reviews":
        submeter_bigquery(df, RAW_TABLE_ORDER_REVIEWS)
    elif nome == "orders":
        submeter_bigquery(df, RAW_TABLE_ORDERS)
    elif nome == "products":
        submeter_bigquery(df, RAW_TABLE_PRODUCTS)
    elif nome == "sellers":
        submeter_bigquery(df, RAW_TABLE_SELLERS)
    elif nome == "category_translation":
        submeter_bigquery(df, RAW_TABLE_CATEGORY_NAME_TRANSLATION)
```
